```python
# ----------------------------------------------pixel operation--------------------------------------------------------

print(img[300,300])

img[300,300] = [255,255,255]
```

```python
# -------------------------------------------------cropping------------------------------------------------------------

img_cropped = img[0:300, 130:325]

plt.imshow(img_cropped)
```

```python
# -----------------------------------------------splitting-------------------------------------------------------------

img_b, img_g, img_r = cv2.split(img)
```

```python
# ----------------------------------------------RGB channels-------------------------------------------------------------

plt.imshow(img_b)
```

```python
plt.imshow(img_g)
```
